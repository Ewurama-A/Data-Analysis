## **Using Siuba and Pandas to analyze itunes data**

Itunes is a popular music steaming service used around the world. The code takes data from an excel file containing the popular tracks, composer, time spent listening to a particular song and the genre as well. 

The goal if this analysis is to process and clean the data, preparing the data for *visulaization*. Each cell explains the code and the processes being done.

```python
#filtering out second index

df = df >> select(~_.a)

df >> _.head()

# using siuba select to check each column for missing data
df.info()
df.shape
df.count()

df >> _.info()
```

###### We can see that data above contains some missing values in the composer's column, hence it will be dealt with.

```python
# filling unknown values in Composer columns
df["Composer"] = df >> select(_.Composer)>> _.fillna("Unknown")

df.info()
```

```python
# Change column name to songtrack and Make track column the index

from siuba import rename

df = df >> rename(Song_Track = _.Track) >> _.head()


# df['Song_Track'] = df >> select(_.Song_Track == _.Track)

# df = df >> select(~ _.Track)

df = df.set_index('Song_Track')

df >> _.head()
```

```python
# Changing unit price name to indicate currency as well
df["Unit_Price_dollars"] = df['UnitPrice']

df.head()

df = df >> select(~_.UnitPrice)

df.head()
```

```python
# Creating minutes column from miiliseonds to minutes to make it comprehensible

df['Total_runtime_min'] = df >> select(_.Milliseconds)

df.head()

# 60000 milliseconds = 1 minute 
df['Total_runtime_min'] = np.divide(df['Total_runtime_min'],60000)

df.head()
```

```python
# Sort out playtime from most played to least played

from siuba import arrange

# using arrange verb to sort playtime in descending order
df = df >> arrange(-_.Total_runtime_min)

df.head()
```

```python
# Changing bytes into megabytes for easier comprehension

df['song_size_Mb'] = df["Bytes"]

df.head()

# Converting bytes to megabytes. 1 Mb = (1024)2 bytes
bytes_megabytes = np.multiply (1024,1024)

df['song_size_Mb'] = np.divide(df['song_size_Mb'], bytes_megabytes)

df.head()
```

```python
# Checking some statistics using numpy

# The mean price of the songs available
np.mean(df['Unit_Price_dollars'])

```

###### About price mean

The average price of a song on this list is $1.39. 

```python
#The mean for size of the songs
np.mean(df['song_size_Mb'])
```

###### About Song size and runtime

The average size of on this list song is 22 megabytes. It seems there are some outliers increasing the song size. 

Upon inspection, it seems there are some episodes on the list meaning that movies,TV episodes and other programs were included in this list.
We used describe() to find the minimum and maximum values and compare to confirm this theory 


We may run into this problem with the total mean runtime as well.

To solve this problem, we could filter out any song track having a runtime greater than 10 minutes, to reflect the 75% percentile's number of 5 minutes. Care must be taken not to casually filter any important data.
The *filter()* verb in ***siuba*** could be used for this

```python
df = df >> filter(_.Total_runtime_min <= 10)

df.head()
```

```python
df.info()
```

```python
# new runtime mean
from siuba.siu import call

df >> call(np.mean, _.Total_runtime_min)

#np.mean(df["Total_runtime_min"])

```

```python
#new mean for songsize

df >> call(np.mean, _.song_size_Mb)

#np.mean(df['song_size_Mb'])
```

#### New averages for runtime and song size

The new mean for the runtime is 4.34 minutes reflecting the average time for a song (Harris,2021). 

The new mean for the song size is 7.8 megabytes which also reflects the average size of a song (Harris,2021).
