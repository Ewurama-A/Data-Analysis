```python
#importing packages to be used

import siuba 
import pandas as pd 
import numpy as np
```

CLEANING DATA WITH PANDAS/SIUBA

This code uses Siuba, Pandas and Numpy to clean,process and save an excel document. The goal is to filter,summarize and group data for further analysis.

Each cell has a comment to explain what is happening.

```python
df = pd.read_csv(r"C:\Users\user\OneDrive\Documents\Uncleaned_DS_jobs.csv")

df.head()
```

```python
from siuba import _,select 

df = df >> select(~_.index)

df >> _.head()

df.info()
#No missing data

#Equivalent in Pandas
#df.drop('index', inplace=True, axis=1)
#df.head()
```

```python
#select salary estimate for cleaning

df >> select(_.contains('Salary Estimate'))

```

```python
df['Salary_Estimate_in_K'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])

df.head()
```

```python
df['Salary_Estimate_in_K'] = df['Salary_Estimate_in_K'].apply(lambda x: x.replace('K','').replace('$',''))

df.head()
```

```python
df['min_salary'] = df['Salary_Estimate_in_K'].apply(lambda x: int(x.split('-')[0]))

df['max_salary'] = df['Salary_Estimate_in_K'].apply(lambda x: int(x.split('-')[1]))

df = df >> select(~_.contains('Salary Estimate'))

df.head()

#remove salary estimate
```

```python
df['avg_salary'] = (df.min_salary+df.max_salary)//2
df.head()
```

```python
df.sort_values(by='Rating', ascending=True)
```

```python
df.Rating = np.where(df.Rating==-1.0,0,df.Rating)

df.sort_values(by='Rating', ascending=True)
```

```python
df['Location'] = df['Location'].apply(lambda x: x.split(',')[-0])

df
```

```python
df['Headquarters'] = df['Headquarters'].apply(lambda x: x.split(',')[-0])

df
```

```python
df['company_age'] = df.Founded.apply(lambda x: x if x<1 else 2023-x)
df['company_age'] = df.company_age.apply(lambda x: x if x>1 else 0)

df
```

```python
df['Job Title'].value_counts()
```

```python
#Change to use regex and siuba

def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
df['seniority'] = df['Job Title'].apply(seniority)

df
```

```python
df['Location'] = df.Location.apply(lambda x: x.replace('United States','US'))

df
```

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
df
```

```python
df.to_csv('Processed_DS_Jobs.csv', index=False)
```
