Which schools have high number of international students?

```python
merged_data.info()


#merged_data.query('stats_pc_intl_students >0').sort_values( by = 'stats_pc_intl_students',ascending= False)
```

```python
# Iterate through column using a for loop and save numbers into new list.

import re
newlist = []
for x in merged_data['stats_pc_intl_students']:
    # Use regex to select only numbers and remove comma(,).
    h = re.findall('\d\d',x)
    # Use join method to combine numbers
    j = ''.join(h)
    newlist.append(j)    

# Replace old column data with new list
merged_data['stats_pc_intl_students'] = pd.Series(newlist)

# fill missing data
# Make new column float type.
merged_data['stats_pc_intl_students'] = merged_data['stats_pc_intl_students'].replace('','0')

merged_data['stats_pc_intl_students'].info()
```

```python
# Make new column float type.
merged_data = merged_data.astype({'stats_pc_intl_students':'float'})

merged_data.query('stats_pc_intl_students >0').sort_values( by = 'stats_pc_intl_students',ascending= False)
```

We see that London Business School, Macau University, and Near East, located in Europe and East Asia are the universities with high number of international students. 

University of Mysore and Wake Forest University are our bottom two with respect to the percentage of international students. They are located in India and United States.
